```python
import pyspark.sql.functions as F
```

```python
print("Initiating automated data ingestion pipeline...")

# Simulate reading raw data from AWS S3 (Raw/Landing Zone)
# In production, this would be: s3://company-datalake/raw/claims/...

# import data
inpatient_path = "dbfs:/FileStore/tables/Train_Inpatientdata.csv"
outpatient_path = "dbfs:/FileStore/tables/Train_Outpatientdata.csv"
beneficiary_path = "dbfs:/FileStore/tables/Train_Beneficiarydata.csv"
```

```python
# load data from CSV to PySpark DataFrame
# inferSchema=True ; automately define the data type (string, int, etc.)

inpatient_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(inpatient_path)
outpatient_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(outpatient_path)
beneficiary_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(beneficiary_path)

# display 5 rows of Inpatient data
print("Inpatient Data Preview:")
display(inpatient_df.limit(5))
```

```python
# Transform and Save data into Delta Format in Bronze Layerฃ

# Write to Databricks Catalog (Bronze Layer)
# This converts raw CSVs into optimized Delta Tables for downstream usage

print("Writing raw data to Bronze Layer (Unity Catalog)...")

bronze_layer_path = "dbfs:/data/bronze/"

print("Saving raw data to Bronze Layer (Delta format)...")
inpatient_df.write.format("delta").mode("overwrite").save(bronze_layer_path + "inpatient")
outpatient_df.write.format("delta").mode("overwrite").save(bronze_layer_path + "outpatient")
beneficiary_df.write.format("delta").mode("overwrite").save(bronze_layer_path + "beneficiary")

print("✅ Data Ingestion to Bronze Layer Completed Successfully!")
```
